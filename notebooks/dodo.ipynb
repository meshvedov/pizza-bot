```python
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50, separators=["## ", "\n\n", "\n", " ", ""])

texts = text_splitter.split_documents(text_docs)
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings_api_model = OpenAIEmbeddings(api_key=openai_api_key, model='text-embedding-3-small', 
                                        base_url="https://api.vsellm.ru/")
```

```python
embeds = embeddings_api_model.embed_documents([t.page_content for t in texts])
```

```python
from langchain_classic.vectorstores import FAISS
vectorstore = FAISS.from_documents(texts, embeddings_api_model)
# text_embedding_pairs = zip(texts, embeds)
# vectorstore = FAISS.from_embeddings(text_embeddings=text_embedding_pairs, embedding=embeddings_api_model)
vectorstore.save_local("dodo_faiss_index")
```

```python
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k":4})
```

```python
def format_docs(docs):
    return "\n\n".join([d.page_content for d in docs])
```

```python
from langchain_core.prompts import FewShotPromptTemplate, PromptTemplate

example_template = """Question: {Question}
Answer: [{Products_name}]"""

example_prompt = PromptTemplate(
    input_variables=["Question", "Products_name"],
    template=example_template,
)

few_shots_prompt = FewShotPromptTemplate(
    examples=few_shots.to_dict(orient="records"),
    example_prompt=example_prompt,
    input_variables=["question"],
    prefix="–¢—ã –ø–æ–º–æ—â–Ω–∏–∫, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–º–æ–≥–∞–µ—Ç –¥–µ–ª–∞—Ç—å –∑–∞–∫–∞–∑—ã –≤ –ø–∏—Ü—Ü–µ—Ä–∏–∏ –î–æ–¥–æ. "
           "–ò—Å–ø–æ–ª—å–∑—É–π —Å–ª–µ–¥—É—é—â–∏–µ –ø—Ä–∏–º–µ—Ä—ã –¥–ª—è –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å:\n\n",
    suffix="\n\nQuestion: {question}\nAnswer:",
    example_separator="\n\n",
)
```

```python
from langchain_classic.chains import ConversationalRetrievalChain
from langchain_classic.memory import ConversationBufferWindowMemory
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

openai_api_key='sk-al8Dzmdmj-xJy13SoUR5aQ'

def get_few_shot_examples(df):
    examples = ""
    for _, row in df.iterrows():
        examples += f"Question: {row['Question']}\nProducts_name: {row['Products_name']}\n"
    return examples

few_shot_examples = get_few_shot_examples(few_shots)


template = f"""
You are a helpful assistant for Dodo Pizza.
You need to understand the user's order and return a list of products based on the provided context.
Use the following examples to understand how to format the output.

{few_shot_examples}

Use the following pieces of context to answer the user's question at the end.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {{context}}
Question: {{question}}
Products_name:
"""

prompt = PromptTemplate(template=template, input_variables=["context", "question"])

memory = ConversationBufferWindowMemory(k=5, memory_key="chat_history", return_messages=True)

embeddings = OpenAIEmbeddings(api_key=openai_api_key, model='text-embedding-3-small', 
                                        base_url="https://api.vsellm.ru/")
db = FAISS.load_local("dodo_faiss_index", embeddings, allow_dangerous_deserialization=True)
retriever = db.as_retriever()

# –î–æ–±–∞–≤–ª—è–µ–º –∫–ª—é—á –∫—É—Ä—Å–∞ –∏ url-–∞–¥—Ä–µ—Å –Ω–∞—à–µ–≥–æ —Å–µ—Ä–≤–µ—Ä–∞
llm = ChatOpenAI(api_key=openai_api_key, model='gpt-4o-mini',
                        base_url="https://api.vsellm.ru/")

qa = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    combine_docs_chain_kwargs={"prompt": prompt},
    verbose=True
)
```
